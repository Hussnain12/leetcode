```python
import os
import csv

def create_increment_csv(filename):
    # Check if the CSV file exists, create if not
    if not os.path.exists(filename):
        with open(filename, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow([0])  # Initial value is 0

    # Read the current value from the CSV
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        current_value = int(next(reader)[0])

    # Increment the value by 1
    new_value = current_value + 1

    # Append the new value to the CSV
    with open(filename, 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([new_value])

    print(f"Updated value: {new_value}")

if __name__ == "__main__":
    csv_filename = "value.csv"
    create_increment_csv(csv_filename)
```
