```python
class Solution:
    def reverseVowels(self, s: str) -> str:
        a='aeiouAEIOU'
        b=''
        for i in s:
            if i in a:
                b=b+i
        c=''
        if len(b)!=0:
            d=b[::-1]
        for i in s:
            if i in a:
                c=c+d[0]
                d=d[1:]
            else:
                c=c+i
        return c
```
